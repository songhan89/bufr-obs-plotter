```python
import pandas as pd
import json
import pdbufr
import glob
```

```python
df = pd.DataFrame()
for f in glob.glob('./input/sample_data/IS*020000'):
    print (f)
    temp_df = pdbufr.read_bufr(f, columns='all',
    filters={'latitude': slice(-15,25),
      'longitude': slice(90,130)},
    flat=True)
    df = pd.concat([df, temp_df], ignore_index=True)
```

```python
variables = ['typicalDate', 'typicalTime', '#1#shipOrMobileLandStationIdentifier', '#1#stationNumber', 
             '#1#stationOrSiteName', '#1#characteristicOfPressureTendency',
             '#1#latitude', '#1#longitude', '#1#pressureReducedToMeanSeaLevel', 
             '#1#airTemperature', '#1#dewpointTemperature', '#1#cloudCoverTotal', 
             '#1#presentWeather', '#1#windSpeed', '#1#windDirection']
df = df[variables]
```

```python
l = []
for c in df.columns:
    l.append(c.split("#")[-1])

df.columns = l
```

```python
import numpy as np

df['u'] = -np.sin(np.radians(df['windDirection'])) * df['windSpeed'] * 1.94384
df['v'] = -np.cos(np.radians(df['windDirection'])) * df['windSpeed'] * 1.94384
```

```python
df
```

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt

from metpy.calc import reduce_point_density
from metpy.cbook import get_test_data
from metpy.io import metar
from metpy.plots import add_metpy_logo, current_weather, sky_cover, StationPlot
```

```python
df = df.fillna(value=np.nan)
```

```python
df['presentWeather'] = df['presentWeather'].astype(float)
```

```python
df['presentWeather'].replace(np.nan, 0, inplace=True)
df['presentWeather'].replace(300, 0, inplace=True)
```

```python
df['presentWeather'] = df['presentWeather'].astype(int)
```

```python
df = df[df['presentWeather'] < 200]
```

```python
df['cloudCoverTotal'].replace(np.nan, 10, inplace=True)
```

```python
df['cloudCoverTotal'] = df['cloudCoverTotal'].values.astype(int) // 12.5
```

```python
df['cloudCoverTotal'] = df['cloudCoverTotal'].astype(int)
```

```python
df['presentWeather'] = 63
```

```python
fig = plt.figure(figsize=(20, 10))
add_metpy_logo(fig, 1100, 300, size='large')
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

# Add some various map elements to the plot to make it recognizable.
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.LAKES)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.STATES)
ax.add_feature(cfeature.BORDERS)

# Set plot bounds
ax.set_extent((90, 130, -5, 25))

stationplot = StationPlot(ax, df['longitude'].values, df['latitude'].values,
                          clip_on=True, transform=ccrs.PlateCarree(), fontsize=12)

# Plot the temperature and dew point to the upper and lower left, respectively, of
# the center point. Each one uses a different color.
stationplot.plot_parameter('NW', df['airTemperature'].values, color='red')
stationplot.plot_parameter('SW', df['dewpointTemperature'].values,
                           color='darkgreen')
stationplot.plot_parameter('NE', df['pressureReducedToMeanSeaLevel'].values,
                           formatter=lambda v: format(v/100, '.0f')[:-3:])
stationplot.plot_symbol('C', df['cloudCoverTotal'] , sky_cover)
stationplot.plot_symbol('W', df['presentWeather'].values, current_weather)
stationplot.plot_barb(df['u'].values, df['v'].values)
```
